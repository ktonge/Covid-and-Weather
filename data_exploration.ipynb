```python
import pandas as pd
import numpy as np

covid_df = pd.read_csv("us_counties_covid19_daily.csv")
covid_df.head()
```

```python
nyc_df = covid_df[(covid_df['state'] == 'New York') & (covid_df['county'] == 'New York City')].reset_index().drop(['index'],1)
nyc_df.iloc[0:30]
```

```python
# Create new columns to hold daily change in cases and deaths

nyc_df['new_cases'] = pd.Series(dtype=int)
nyc_df['new_deaths'] = pd.Series(dtype=int)

nyc_df.head()
```

```python
# Record values for daily change in cases and deaths
for i in range(1,len(nyc_df.index)):
    
    # Set new cases
    current_cases = nyc_df.iloc[i,4]
    yesterday_cases = nyc_df.iloc[i-1,4]
    nyc_df.iloc[i,6] = current_cases - yesterday_cases

    # Set new deaths
    current_deaths = nyc_df.iloc[i,5]
    yesterday_deaths = nyc_df.iloc[i-1,5]
    nyc_df.iloc[i,7] = current_deaths - yesterday_deaths

nyc_df.head(30)
```

```python
# Create columns to hold 7 and 14 day changes in cases and deats
nyc_df['case_delta_7'] = pd.Series(dtype=int)
nyc_df['death_delta_7'] = pd.Series(dtype=int)
nyc_df['case_delta_14'] = pd.Series(dtype=int)
nyc_df['death_delta_14'] = pd.Series(dtype=int)

nyc_df.head()
```

```python
# Record the change in number of cases and deaths from 7 days ago to today
for i in range(7,len(nyc_df.index)):

    previous_cases = nyc_df.iloc[i-7,4]
    current_cases = nyc_df.iloc[i,4]
    previous_deaths = nyc_df.iloc[i-7,5]
    current_deaths = nyc_df.iloc[i,5]
    
    nyc_df.iloc[i,8] = current_cases - previous_cases
    nyc_df.iloc[i,9] = current_deaths - previous_deaths

# Record the change in number of cases and deaths from 14 days ago to today
for i in range(14,len(nyc_df.index)):
    previous_cases = nyc_df.iloc[i-14,4]
    current_cases = nyc_df.iloc[i,4]
    previous_deaths = nyc_df.iloc[i-14,5]
    current_deaths = nyc_df.iloc[i,5]

    nyc_df.iloc[i,10] = current_cases - previous_cases
    nyc_df.iloc[i,11] = current_deaths - previous_deaths

nyc_df.head(30)
```

```python
nyc_df.to_csv('nyc_covid_all_counts.csv')
```
